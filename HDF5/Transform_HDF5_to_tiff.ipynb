```python
import h5py
import numpy as np
import os

# === 入力ファイルと出力ファイルのパス ===
input_hdf5_path = r"\\4-307-km\p\Tardigrade\Tardigrada_mukumoto\orig\raw_t=0~225\raw_t=1~225.h5"
output_hdf5_part1 = r"H:\kumamushi-new\split_part1_t000-112.h5"
output_hdf5_part2 = r"H:\kumamushi-new\split_part1_t113-224.h5"

# 分割範囲
split_ranges = {
    output_hdf5_part1: (0, 113),    # t = 0~112
    output_hdf5_part2: (113, 225)   # t = 113~224
}

# データを1フレームずつコピー
with h5py.File(input_hdf5_path, 'r') as f_in:
    dataset_name = list(f_in.keys())[0]
    data = f_in[dataset_name]
    z, c, t_total, y, x = data.shape
    dtype = data.dtype

    for out_path, (t_start, t_end) in split_ranges.items():
        t_len = t_end - t_start
        print(f"Writing {out_path} ...")

        with h5py.File(out_path, 'w') as f_out:
            dset = f_out.create_dataset(
                dataset_name,
                shape=(z, c, t_len, y, x),
                dtype=dtype,
                chunks=(z, c, 1, y, x)  # 時間方向1ボリュームごとにチャンク化
    
        
            )

            for i, t in enumerate(range(t_start, t_end)):
                print(f"  Copying t={t} → t'={i}")
                dset[:, :, i, :, :] = data[:, :, t, :, :]
```
