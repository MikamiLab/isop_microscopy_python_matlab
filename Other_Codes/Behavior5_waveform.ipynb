```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data
data = pd.read_csv('tmac_activities_precorrected_alllabels_discard50_0807_behavior5_6_BehavType.csv')

# Extract Behavior5 signal
behavior5 = data['Behavior5']

# Convert index to time assuming 10 Hz (1 volume = 0.1 s)
time_seconds_10Hz = data.index * 0.1

# Define plot range: 56s to 66s -> vol 560 to 660
start_vol = 560
end_vol = 660
plot_times = time_seconds_10Hz[start_vol:end_vol + 1]
plot_values = behavior5[start_vol:end_vol + 1]

# Define volumes to mark
highlight_vols = [596, 601, 606, 611, 616, 621, 626, 631]
highlight_times = [v * 0.1 for v in highlight_vols]
highlight_values = behavior5.iloc[highlight_vols]

# Define target aspect ratio
width = 14.3
height = 2.55

# Create plot with specified aspect ratio
plt.figure(figsize=(width, height))
plt.plot(plot_times, plot_values, color='black', linewidth=1.0, label='Behavior5')
plt.scatter(highlight_times, highlight_values, color='magenta', zorder=3, label='Highlighted Points')

# Styling
plt.title('Behavior5 waveform (56–66 s at 10 Hz)')
plt.xlabel('Time [s]')
plt.ylabel('Behavior5')
plt.xticks(ticks=range(56, 67, 1))  # 56 to 66 seconds in 1-second steps
plt.grid(False)
plt.tight_layout()

# Save and show
plt.savefig('behavior5_waveform_56-66s_10Hz.png', dpi=300, bbox_inches='tight')
plt.show()
```
