```python
# 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 
file_path = "Vent_Dor_timepont_Behavior5.xlsx"
df = pd.read_excel(file_path)

# 
motif5 = df['Behavior5']
ventral_mask = df['Ventral'] == 1
dorsal_mask = df['Dorsal'] == 1

#   10 Hz
time_sec = df.index / 10.0

# : 36 : 62  figsize = 6.2  3.6
plt.figure(figsize=(6.2, 3.6))

# motif #5 
plt.plot(time_sec, motif5, color='gray', linewidth=1.5)

# 
plt.scatter(time_sec[ventral_mask], motif5[ventral_mask],
            color='mediumseagreen', s=40, marker='o', zorder=3)
plt.scatter(time_sec[dorsal_mask], motif5[dorsal_mask],
            color='mediumpurple', s=40, marker='o', zorder=3)

# 
plt.xlabel("Time (s)", fontsize=10)
plt.ylabel("Motif #5 amplitude", fontsize=10)
plt.title("Motif #5 time series with annotated head bends", fontsize=12)

# X20
max_time = time_sec.max()
plt.xticks(np.arange(0, max_time + 1, 20), fontsize=10)
plt.yticks(fontsize=10)

# 
plt.grid(False)
plt.tight_layout()

plt.savefig('Motif #5 time series with annotated head bends', dpi=300, bbox_inches='tight')

# 
plt.show()
```
